```python
import os
CATALOG_NAME = "proyecto_final"
SCHEMA_NAME = "landing"
VOLUME_NAME = "raw_data"
PROJECT_NAME = 'dmc_final'
VOLUME_PATH = f"Volumes/{CATALOG_NAME}/{SCHEMA_NAME}/{VOLUME_NAME}/{PROJECT_NAME}"
spark.sql(f"create catalog if not exists {CATALOG_NAME}")
spark.sql(f"create schema if not exists {CATALOG_NAME}.{SCHEMA_NAME}")
spark.sql(f"create schema if not exists {CATALOG_NAME}.bronze")
spark.sql(f"create schema if not exists {CATALOG_NAME}.silver")
spark.sql(f"create schema if not exists {CATALOG_NAME}.gold")
spark.sql(f"create schema if not exists {CATALOG_NAME}.bronze_dev")
spark.sql(f"create schema if not exists {CATALOG_NAME}.silver_dev")
spark.sql(f"create schema if not exists {CATALOG_NAME}.gold_dev")
spark.sql(f"create volume if not exists {CATALOG_NAME}.{SCHEMA_NAME}.{VOLUME_NAME}")
```

```python
dbutils.fs.mkdirs("/Volumes/proyecto_final/landing/raw_data/dmc_final/clientes")
dbutils.fs.mkdirs("/Volumes/proyecto_final/landing/raw_data/dmc_final/productos")
dbutils.fs.mkdirs("/Volumes/proyecto_final/landing/raw_data/dmc_final/pedidos")
dbutils.fs.mkdirs("/Volumes/proyecto_final/landing/raw_data/dmc_final/detalle_pedidos")
dbutils.fs.mkdirs("/Volumes/proyecto_final/landing/raw_data/_schemas/dmc_final/clientes")
dbutils.fs.mkdirs("/Volumes/proyecto_final/landing/raw_data/_schemas/dmc_final/productos")
dbutils.fs.mkdirs("/Volumes/proyecto_final/landing/raw_data/_schemas/dmc_final/pedidos")
dbutils.fs.mkdirs("/Volumes/proyecto_final/landing/raw_data/_schemas/dmc_final/detalle_pedidos")
```

```python
clients_f = ['clientes_batch_1.csv', 'clientes_batch_2.csv', 'clientes_batch_3.csv']
products_f = ['productos_batch_1.csv', 'productos_batch_2.csv', 'productos_batch_3.csv']
orders_f = ['pedidos_batch_1.json', 'pedidos_batch_2.json', 'pedidos_batch_3.json']
order_details_f = ['detalle_pedidos_batch_1.json', 'detalle_pedidos_batch_2.json', 'detalle_pedidos_batch_3.json']
to_upload = []
def expected_files(folder, files):
    ret = []
    for f in files:
        path_f = f"/{VOLUME_PATH}/{folder}/{f}"
        if not os.path.exists(path_f):
            ret.append(path_f)
    return ret        
to_upload.extend(expected_files('clientes',clients_f))
to_upload.extend(expected_files('productos',products_f))
to_upload.extend(expected_files('pedidos',orders_f))
to_upload.extend(expected_files('detalle_pedidos',order_details_f))
if to_upload:
    raise FileNotFoundError(
        "Archivos no disponibles\n" +
        "Subir los siguientes archivos antes de continuar:\n" +
        "\n".join(to_upload)
    )
```
